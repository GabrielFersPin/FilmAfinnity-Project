# Data processing.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
```

```python
movies["name"] = movies["actors"].str.split(",")
movies
```

```python
movies = movies.explode("name").drop(columns="actors")
movies["name"] = movies["name"].str.strip()
```

```python
movies = movies.reset_index(drop=True)

movies
```
